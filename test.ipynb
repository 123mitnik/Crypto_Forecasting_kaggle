## Before submisson

```python
#crypto order at same timestamp in the aftergame train data
data_path = './data'
sup_train=pd.read_csv(data_path+'/supplemental_train.csv')
assets_order = dict((t,i) for i,t in enumerate(sup_train.Asset_ID[:N_ASSETS]))
features = ['Count', 'Open', 'High', 'Low', 'Close', 'Volume', 'VWAP',
       'Upper_Shadow', 'Lower_Shadow', 'spread', 'mean_trade',
       'log_price_change', 'asset_order']

#first 15mins samples(15*14=210 rows)
placeholder = get_features(sup_train[:WINDOW_SIZE * (N_ASSETS)])
pd.options.mode.chained_assignment = None  # default='warn'
placeholder['asset_order'] = placeholder.Asset_ID.map(assets_order)
print(placeholder.shape)
# shape into our model's one sample of 15minutes input array
test_x = np.array(placeholder[features])
test_x = test_x.reshape(-1, (N_ASSETS), test_x.shape[-1])
print(test_x.shape)
test_x = np.expand_dims(test_x, axis=0) #additional dimension for one sample in a batch
```

```python
##asset order in the API
example = pd.read_csv(data_path+'/example_test.csv')[:WINDOW_SIZE - 1]
example['asset_order'] = example.Asset_ID.map(assets_order) 
#help adding asset_order to the API delivery
example = example[['Asset_ID','asset_order']]
```

## Start submission to all API delivery

```python
target_pred = []
submission = []
for i in range(4):
    print(i)
    #one minute API sample
    test_csv,sample_prediction_df_csv = f'./data/temp/{i}test_df.csv', f'./data/temp/{i}sample_prediction_df.csv'
    test_df, sample_prediction_df= pd.read_csv(test_csv,usecols =range(1,11)),pd.read_csv(sample_prediction_df_csv,usecols =range(1,3))
    #add features
    test_df = get_features(test_df)
    #sort the test API with the training structure
    test_data = test_df.merge(example, how='outer', on='Asset_ID').sort_values('asset_order')
    test = np.array(test_data[features].fillna(0))#14x13
    test = test.reshape(-1, 1, N_ASSETS, test.shape[-1])#add two more axis:batch, minute
    ## merge with the last 14minute in test_x to form a 15minute test input, and rolling forward
    test_x = np.hstack([test_x, test])[:,-1 * WINDOW_SIZE:]
    ##ready for prediction
    test_data['Target'] = model.predict(test_x).squeeze()
    target_pred.append(test_data)
    ##add to submission, match with row_id
    for _, row in test_df.iterrows():
        try: 
            sample_prediction_df.loc[sample_prediction_df['row_id'] == row['row_id'], 'Target'] = test_data.loc[test_data['row_id'] == row['row_id'], 'Target'].item()
        except: 
            sample_prediction_df.loc[sample_prediction_df['row_id'] == row['row_id'], 'Target'] = 0
    #env.predict(sample_prediction_df)
    submission.append(sample_prediction_df)

print(pd.concat(submission,axis=1))
target_pred = pd.concat(target_pred,axis=0)
```

## Score: weighted correlation

```python
weights= pd.read_csv('./data/asset_details.csv')
```

```python
target_true = pd.read_csv('./data/supplemental_train.csv',nrows=4*14)[['timestamp','Asset_ID','Target']]
target_pred = target_pred[['timestamp','Asset_ID','Target']]
score_df = target_pred.merge(target_true,suffixes=('_pred', '_true'),how='left', on=['timestamp','Asset_ID'])
```

```python
score_df = pd.pivot_table(score_df,index=['Asset_ID','timestamp'])
score_public = 0
ww = 0
for id in range(14):
    s = score_df.loc[id].corr().iloc[0,1]
    w = weights.loc[weights['Asset_ID']==id,'Weight'].item()
    ww = ww+w
    score_public = score_public + s*w

print(score_public/ww)
```
